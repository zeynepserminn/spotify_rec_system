```python
from sklearn.preprocessing import OneHotEncoder

df=data
# OneHotEncoder'ı başlat, sparse=False kullanarak dense bir matris oluşturmayı seçiyoruz
encoder = OneHotEncoder(sparse_output=False)

# album_name sütunu için encoding işlemi
album_encoded = encoder.fit_transform(df[['album_name']])

# OneHotEncoder ile sütun isimlerini alma
column_names = encoder.get_feature_names_out(['album_name'])

# Yeni dataframe'e çevir
album_encoded_df = pd.DataFrame(album_encoded, columns=column_names)

# Orijinal veri setine birleştirme (isteğe bağlı)
df_encoded = pd.concat([df, album_encoded_df], axis=1)

# İlk birkaç satırı gör
print(df_encoded.head(10))
```
