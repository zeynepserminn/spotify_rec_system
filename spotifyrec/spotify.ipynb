```python
import pandas as pd
data = pd.read_csv('dataset.csv')
data.head()



```

```python
import re
data=data.dropna()
data.head()
```

```python

data['album_name'] = data['album_name'].apply(lambda x: re.sub(r';', ' ', x))
data['track_name'] = data['track_name'].apply(lambda x: re.sub(r';', ' ', x))
data.head(15)
```

```python

data['album_name'] = data['album_name'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
data['track_name'] = data['track_name'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
data.head(15)
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
data['artists'] = data['artists'].str.replace(';', ',', regex=False)

data.head(15)


```

```python
data['artists'] = data['artists'].fillna('')
data['artists'] = data['artists'].astype(str).str.split(',')

data.head(15)

```
